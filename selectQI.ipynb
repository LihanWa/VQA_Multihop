```python
import nltk
from nltk import pos_tag
from nltk.tokenize import word_tokenize

# 下载必要的nltk数据包
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

common_adjectives = {"left", "right", "up", "down", "red", "green", "blue", "big", "small"}

def check_nouns(sentence):
    # 分词
    words = word_tokenize(sentence)
    # 词性标注
    tagged_words = pos_tag(words)
    # 打印词性标注的结果
    # print(tagged_words)
    # 统计名词数量，但排除在形容词词典中的词
    noun_count = sum(1 for word, pos in tagged_words if pos in ['NN', 'NNS', 'NNP', 'NNPS'] and word.lower() not in common_adjectives)

    return noun_count > 1


# 示例句子
sentence = "Is the candle to the right of a child?"
if check_nouns(sentence):
    print("这个句子有两个以上的名词。")
else:
    print("这个句子没有两个以上的名词。")
```
